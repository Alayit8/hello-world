**Problem 13**

**Concept:** *Extracting sublists using slicing*

**Task:**

From the list:

* Extract first three elements and assign it to a variable `first_three`

* Extract last three elements and assign it to a variable `last_three`

* Combine them into a new list and assign it to a variable `combined`

* Compute the mean of the combined list and assign it to a variable `mean_combined`

* Print the combined list and the mean

**Notes:**

* You may use the built-in functions `len` and `sum`; you may not
use any loops or library functions.

**Sample Output:**

```
The first 3 elements and last 3 elements combined are : [1, 4, 5, 3, 6, 7]
And the mean of the combined list is 5.0
```

````python
# starter code:
values = [6,1,7,3,5,8,2,3,6,1]

# TO DO: WRITE YOUR CODE/SOLUTION BELOW

first_three = values[:3]
last_three = values[-3:]
combined = first_three + last_three

mean_combined = sum(combined) / len(combined)

print("The first 3 elements and last 3 elements combined are:", combined)
print("And the mean of the combined list is", mean_combined)
````
